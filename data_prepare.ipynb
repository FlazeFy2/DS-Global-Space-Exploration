```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global_Space_Exploration_Dataset.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
 
# Data Preparation - Handle Missing Value
df.isnull().sum()

# Data Preparation - Invalid Numeric Column
ratio_column = [
    'Year','Budget (in Billion $)','Success Rate (%)','Duration (in Days)'
]
for dt in ratio_column:
    plt.boxplot(df[dt])
    plt.title(f'Boxplot of {dt}')
    plt.show()
    
    invalid = df[df[dt] < 0]
    print(invalid.index)
```
