```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global_Space_Exploration_Dataset.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Mission Type','Satellite Type','Environmental Impact']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"{dt} Comparison")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global_Space_Exploration_Dataset.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Collaboration Comparison Space Exploration
df["Country Count Type"] = df["Collaborating Countries"].apply(lambda x: "Multilateral" if pd.notna(x) and "," in x else "Bilateral")
top_context = df["Country Count Type"].value_counts()

# Pie Chart
plt.figure(figsize=(8, 8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Collaboration Comparison Space Exploration")
plt.ylabel("")  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Mission Duration Category (Days)
df = pd.read_csv("Global_Space_Exploration_Dataset.csv")

# Convert duration
def categorize_duration(time):
    if time <= 7:
        return "A Week"
    elif time <= 30:
        return "A Month"
    elif time <= 180:
        return "Half a Year"
    elif time <= 360:
        return "A Year"
    else:
        return "More than A Year"

df["Mission Duration Category"] = df["Duration (in Days)"].apply(categorize_duration)

# Count the number of missions
quartal_counts = df["Mission Duration Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Mission Duration Category (Days)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
